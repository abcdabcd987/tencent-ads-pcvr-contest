```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
embed_percentiles = [[[] for i in range(390)] for j in range(3)]
linear_percentile = [[] for i in range(3)]
product_percentile = [[] for i in range(3)]
mixed_percentile = [[] for i in range(3)]

files = ['../log/Criteo/PNN2/04-24-12:20:42/percentile_15/events.out.tfevents.1493007645.Altria',
        '../log/Criteo/PNN2/04-24-12:20:42/percentile_50/events.out.tfevents.1493007646.Altria',
        '../log/Criteo/PNN2/04-24-12:20:42/percentile_85/events.out.tfevents.1493007646.Altria',]

for k in range(3):
    cnt = 0
    for e in tf.train.summary_iterator(files[k]):
        for v in e.summary.value:
            if 'embed' in v.tag:
                idx = int(v.tag.split('_')[1][4:])
                embed_percentiles[k][idx].append(v.simple_value)
            elif v.tag == 'linear_node0_percentile':
                linear_percentile[k].append(v.simple_value)
            elif v.tag == 'product_node0_percentile':
                product_percentile[k].append(v.simple_value)
            elif v.tag == 'mixed_node0_percentile':
                mixed_percentile[k].append(v.simple_value)
            cnt += 1
            if cnt % 1000000 == 0:
                print cnt
```

```python
embed_percent = np.array(embed_percentiles)
linear_percent = np.array(linear_percentile)
product_percent = np.array(product_percentile)
mixed_percent = np.array(mixed_percentile)

print embed_percent.shape
print linear_percent.shape
print product_percent.shape
print mixed_percent.shape
```

```python
del embed_percentiles, linear_percentile, product_percentile, mixed_percentile
```
